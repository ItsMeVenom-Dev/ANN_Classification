```python
import pandas as pd
data=pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data.head()
```

### Encoding Categorical Variable


```python
from sklearn.preprocessing import StandardScaler, LabelEncoder


label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data["Gender"])

```

```python
data
```

### Applying one hot Encoder in Geographical Column


```python
from sklearn.preprocessing import OneHotEncoder

oneHotEncoder_Geo=OneHotEncoder()
geo_Encoder=oneHotEncoder_Geo.fit_transform(data[['Geography']])
```

```python
oneHotEncoder_Geo.get_feature_names_out(['Geography'])
```

```python
geo_Encoder=geo_Encoder.toarray()
```

```python
geo_encoded_df=pd.DataFrame(geo_Encoder,columns=oneHotEncoder_Geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
data=data.drop(columns='Geography', axis=1)
```

```python

data=pd.concat([data,geo_encoded_df],axis=1)
```

```python
data.head()
```

```python
import pickle
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_georaphy.pkl','wb') as file:
    pickle.dump(oneHotEncoder_Geo,file)
```

```python
data.head()
```

### divide the dataset into dependent and independent feature 

```python
X=data.drop('Exited',axis=1)
Y=data['Exited']
```

### spliting the data in traning and testing sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
with open("scaler.pkl","wb") as file:
    pickle.dump(scaler,file)
```

### ANN Implementation


```python
X_train.shape[1]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(16,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
##to implement optimizer, loss

import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
### compile the model

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#set up the tensorBoard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

log_dir="logs/dir/"+datetime.datetime.now().strftime("%d%m%Y-%H%M%S")
tensorflow_callback=TensorBoard(log_dir,histogram_freq=1)
```

```python
earlystopping_callback=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(
    X_train,y_train,
    validation_data=(X_test,y_test),
    epochs=100,
    callbacks=[tensorflow_callback,earlystopping_callback]
)
```

```python
model.save('model.h5')
```
